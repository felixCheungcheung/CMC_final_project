```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
anno_df = pd.read_csv('lines_data.csv')
banshi_df = anno_df.loc[-anno_df['Banshi'].isnull()]
# banshi_df = anno_df['Banshi'].isnull()
```

```python
banshi_df   
```

```python
banshi_df['Banshi'].unique()
res = {}
for banshi in banshi_df['Banshi'].unique():

    res[banshi] = banshi_df.loc[banshi_df['Banshi'].values==banshi].shape[0]
    # print(f'{banshi} with {res[banshi]}')

plt.rcdefaults()
fig, ax = plt.subplots(figsize=(10,12))

y, x = list(res.keys()), list(res.values())
ax.barh(y,x)
print(y)
ax.set_xlabel('Number of Banshi')
ax.set_title('Banshi Distribution')
plt.show()
```
